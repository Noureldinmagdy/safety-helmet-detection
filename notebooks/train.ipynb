```python
from ultralytics import YOLO
import optuna
import yaml
```

```python
with open('params.yaml', 'r') as file:
    prime_service = yaml.safe_load(file)
```

```python
prime_service
```

```python
def objective(trial):

    lr = round(trial.suggest_float("lr0", float(prime_service["lr"]["low"]), float(prime_service["lr"]["high"])), 2)
    dropout = round(trial.suggest_float("dropout", prime_service["dropout"]["low"], prime_service["dropout"]["high"]), 1)
    batch = trial.suggest_categorical("batch", prime_service["batch_size"])

    model = YOLO("yolo11n.pt") 

    results = model.train(
        data="safety-helmet-1/data.yaml",
        imgsz=128,
        epochs=10,
        batch=batch,
        lr0=lr,
        dropout=dropout,
        project="runs",
        name=f"optuna_trial_{trial.number}",
        verbose=False
    )
    model.export(format="onnx")

    metrics = model.val(name=f"eval_optuna_trial_{trial.number}")
    return metrics.box.map50  

# Create Optuna study
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=3)

# Best result
print("Best trial:")
print(study.best_trial.params)
```
